```python
import pandas as pd

# Open data
df_data = pd.read_parquet("data\cleaned_maintenance_data.parquet")
# Set dtypes
types = {
    "hour": int,
    "day": int,
    "month": int,
    "year": int,
    "is_renting": bool,
    "is_returning": bool,
    "is_installed": bool,
}
df_data = df_data.astype(types)
df_data
```

```python
# Load station info
df_stations = pd.read_csv("data\Informacio_Estacions_Bicing.csv")
df_stations
```

```python
# Merge location data
df_data = pd.merge(df_data, df_stations[["station_id","name","lat","lon","altitude", "capacity", "is_charging_station"]], on="station_id", how="inner")
```

```python
df_data["datetime"] = pd.to_datetime(df_data[["year", "month", "day", "hour"]])
df_data
```

```python
import plotly.express as px

# Create a dataframe with unique datetime values
unique_times = df_data['datetime'].sort_values().unique()
# Extract unique values for sliders
unique_years = df_data['year'].astype(int).unique()
unique_years = list([int(y) for y in unique_years])
unique_months = list(range(1, 13))
unique_days = list(range(1, 32))
unique_hours = list(range(24))
```

```python
from dash import Dash, dcc, html, Input, Output, State
import pandas as pd
import plotly.express as px


app = Dash(__name__)

app.layout = html.Div([
    html.H1("Bike Station Status Over Time"),
    dcc.Graph(id='map', animate=True),
    dcc.Store(id='map-view', data={'lat': 41.3999, 'lon': 2.17162, 'zoom': 11.5}),
    html.Div([
        html.Label("Select Year:"),
        dcc.Slider(
            id='year-slider',
            min=min(unique_years),
            max=max(unique_years),
            value=2020,
            marks={year: str(year) for year in unique_years},
            step=1
        ),
        html.Label("Select Month:"),
        dcc.Slider(
            id='month-slider',
            min=min(unique_months),
            max=max(unique_months),
            value=min(unique_months),
            marks={month: str(month) for month in unique_months},
            step=1
        ),
        html.Label("Select Day:"),
        dcc.Slider(
            id='day-slider',
            min=min(unique_days),
            max=max(unique_days),
            value=min(unique_days),
            marks={day: str(day) for day in unique_days if day % 5 == 0},
            step=1
        ),
        html.Label("Select Hour:"),
        dcc.Slider(
            id='hour-slider',
            min=min(unique_hours),
            max=max(unique_hours),
            value=min(unique_hours),
            marks={hour: str(hour) for hour in unique_hours},
            step=1
        )
    ])
])

@app.callback(
    Output('map', 'figure'),
    Output('map-view', 'data'),
    Input('year-slider', 'value'),
    Input('month-slider', 'value'),
    Input('day-slider', 'value'),
    Input('hour-slider', 'value'),
    State('map', 'relayoutData'),
    State('map-view', 'data')
)
def update_map(selected_year, selected_month, selected_day, selected_hour, relayout_data, map_view):
    # Ensure the selected date and time is valid
    try:
        selected_time = pd.Timestamp(year=selected_year, month=selected_month, day=selected_day, hour=selected_hour)
        filtered_data = df_data[df_data['datetime'] == selected_time]
    except ValueError:
        filtered_data = pd.DataFrame(columns=df_data.columns)  # empty DataFrame if date is invalid

    # Check if relayout_data contains new map view settings
    if relayout_data and 'mapbox.center' in relayout_data:
        map_view['lat'] = relayout_data['mapbox.center']['lat']
        map_view['lon'] = relayout_data['mapbox.center']['lon']
        map_view['zoom'] = relayout_data['mapbox.zoom']

    fig = px.scatter_mapbox(
        filtered_data, 
        lat="lat", 
        lon="lon", 
        hover_name="name", 
        hover_data=["status", "num_bikes_available"],
        color='status',
        color_discrete_map={'IN_SERVICE': 'green', 'NOT_IN_SERVICE': 'red', "MAINTENANCE": "gray", "PLANNED": "blue"},
        zoom=map_view['zoom'], 
        height=600,
        center={"lat": map_view['lat'], "lon": map_view['lon']}
    )

    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    return fig, map_view

if __name__ == '__main__':
    app.run(debug=True)
# To see it better, open a new tab in your browser and go to http://localhost:8050/
```
